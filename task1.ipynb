## Task 1 : Prediction Using Suupervised Machine Learning
## GRIP @ The Sparks Foundation 
Predict the percentage of an student based on the no. of study hours, involves only two variables. 

```python
#importing the required libraries
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt 
import pandas as pd 
import numpy as np 
```

## Step 1 : Reading the data from source

```python
url = r"https://raw.githubusercontent.com/AdiPersonalWorks/Random/master/student_scores%20-%20student_scores.csv"
s_data = pd.read_csv(url)
print("Data import successful")

s_data.head(10)
```

## Step 2 : Input Data Visualization

```python
s_data.plot(x="Hours",y="Scores",style='o')
plt.title("hours vs percentage")
plt.xlabel("hours studied")
plt.ylabel("percentage score")
plt.show()
```

## Step 3 : Data Preprocessing
Division of data in attributes and labels


```python
X = s_data.iloc[:,:-1].values
y = s_data.iloc[:,1].values
```

## Step 4 : Model Training
splitting data in training and testing data sets


```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
regressor = LinearRegression()
regressor.fit(X_train.reshape(-1,1),y_train)

print("Training complete.")
```

## Step 5 : Plotting the line of regression
time to visualize the best fit line

```python
#plotting regression line
line = regressor.coef_*X+regressor.intercept_

#plotting for test data
plt.scatter(X,y)
plt.plot(X, line, color='red')
plt.show()
```

## Step 6 : Making Predictions


```python
#testing data
print(X_test)
#model prediction
y_pred = regressor.predict(X_test)
```

## Step 7 : Comparing Actual result to the predicted model result

```python
df = pd.DataFrame({'Actual': y_test, 'predicted': y_pred})
df
```

```python
#estimating training and test score
print("training score:",regressor.score(X_train,y_train))
print("test score:",regressor.score(X_test,y_test))
```

```python
#plotting the bar graph to see the difference between actual and predicted value

df.plot(kind='bar',figsize=(5,5))
plt.grid(which='major', linewidth='0.5', color='red')
plt.grid(which='minor', linewidth='0.5', color='blue')
plt.show()
```

```python
# TESTING MODEL WITH OWN DATA
hours = 9.25
test = np.array([hours])
test = test.reshape(-1,1)
own_pred = regressor.predict(test)
print("No. of hours = {}".format(hours))
print("Predicted score = {}".format(own_pred[0]))
```

## step 8 : Evaluating the model

```python
from sklearn import metrics
print('mean absolute error: ', metrics.mean_absolute_error(y_test, y_pred))
print('mean squared error: ', metrics.mean_squared_error(y_test, y_pred))
print('root mean squared error: ', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
print('R-2: ', metrics.r2_score(y_test, y_pred))
```

R-2 gives the score of model fit and in this case R-2 = 0.9454906892105355 which is good
